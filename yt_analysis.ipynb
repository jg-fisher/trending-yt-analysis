```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings('ignore')

data = pd.read_csv('USvideos.csv')
```

```python
likes_m = np.mean(data['likes'])
dislikes_m = np.mean(data['dislikes'])

fig = plt.figure()
plt.title('Average Dislikes vs Likes')
s = fig.add_subplot(111)
s.bar(['Avg. Likes', 'Avg. Dislikes'], [likes_m, dislikes_m])
```

```python
title_lengths = [len(title.split(' ')) for title in data['title']]
ax = plt.axes()
ax.set_title('Title Word Length')
title_length_dist = sns.distplot(title_lengths, ax=ax)
```

```python
descriptions = data['description'].astype('str')
description_lengths = [len(d.split(' ')) for d in descriptions]
ax = plt.axes()
ax.set_title('Description Word Length')
description_length_dist = sns.distplot(description_lengths)
```

```python
descriptions = data['description'].astype('str')
title_char_lengths = [len(d) for d in descriptions]
ax = plt.axes()
ax.set_title('Title Character Length')
title_length_dist = sns.distplot(title_char_lengths, ax=ax)
```

```python
descriptions = data['description'].astype('str')
description_char_lengths = [len(d) for d in descriptions]
ax = plt.axes()
ax.set_title('Description Character Length')
description_length_dist = sns.distplot(description_char_lengths)
```
