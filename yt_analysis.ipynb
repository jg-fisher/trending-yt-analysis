```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('USvideos.csv')
data.columns.values
```

```python
likes_m = np.mean(data['likes'])
dislikes_m = np.mean(data['dislikes'])

fig = plt.figure()
s = fig.add_subplot(111)
s.bar(['Avg. Likes', 'Avg. Dislikes'], [likes_m, dislikes_m])
```

```python
title_lengths = [len(title) for title in data['title']]
np.mean(title_lengths)
```

```python
ax = plt.axes()
ax.set_title('Title Length')

title_length_dist = sns.distplot(title_lengths, ax=ax)
```

```python
descriptions = data['description'].astype('str')
description_lengths = [len(d) for d in descriptions]

ax = plt.axes()
ax.set_title('Description Length')

description_length_dist = sns.distplot(description_lengths)
```
